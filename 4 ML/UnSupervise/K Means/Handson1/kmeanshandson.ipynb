```python
""" How to choose the number of clusters
However choosing the right value of of K can be tricky?

If you choose too few clusters than you might miss some important patterns in the data
If you choose too many clusters than the model might become too complex and overfit the data

Small datasets: For small datasets, you might try k-values from 1 to a value 10 or 15.

Larger datasets: For larger datasets, you can explore a wider range
"""
```

```python
from sklearn.cluster import KMeans
```

```python
k_values = [2,3,4,5,6,7]
```

```python
# elbow

ssd=[]

for k in k_values:
  km = KMeans(n_clusters=k, max_iter=150,random_state=32)
  km.fit(df_scaled)
  ssd.append(km.inertia_)
```

```python
plt.plot(k_values,ssd)
```

```python
from sklearn.metrics import silhouette_score

k_values = range(2,8)

silhouette_scores = []
for k in k_values:
  kmeans = KMeans(n_clusters=k,random_state=32)
  kmeans.fit(df_scaled)
  silhouette_avg = silhouette_score(df_scaled, kmeans.labels_)
  silhouette_scores.append(silhouette_avg)

plt.subplot(1, 2, 2)
plt.plot(k_values, silhouette_scores, marker='o', color='green')
plt.title('Silhouette Score vs. k')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Silhouette Score')
```

```python
# 4 will be considered as number of clusters as 4 has sharper bend between 4 and 5, 4 has high silhoutte score than 5
```

```python
kmodel = KMeans(n_clusters=4, max_iter=150, random_state=32)
kmodel.fit(df_scaled)
```

```python
print(kmodel.labels_)  # this will assign each data point into 4 different clusters
```

```python
df['Cluster_ID']=kmodel.labels_
```
