```python
import pandas as pd
```

```python
df=pd.read_csv(r'/content/cars-1 (2).csv')
```

```python
df
```

```python
df.info()
```

```python
df['cubicinches'].unique()
```

```python
#object to numeric:to_numeric
#errors='coerce'-> converts all the error to null value
df['cubicinches']=pd.to_numeric(df['cubicinches'],errors='coerce')
```

```python
df.isnull().sum()
```

```python
df['weightlbs'].unique()
```

```python
df['weightlbs']=pd.to_numeric(df['weightlbs'],errors='coerce')
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.info()
```

```python
#subplots
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.subplots(nrows=2,ncols=2)
plt.show()
```

```python
# 1. Create scatter plots for the following data, make sure all the plots appear in the same
# plane with x-labels and y-labels evenly spaced.
# a. Create scatter plots with between (mpg-hp, mpg-weight), and (hp-mpg,
# hp-time-to-60) with respect to each brand.
```

```python
fig,axis=plt.subplots(2,2,figsize=(16,8))
sns.scatterplot(data=df,x='mpg',y='hp',ax=axis[0,0],hue='brand')
sns.scatterplot(data=df,x='mpg',y='weightlbs',ax=axis[0,1],hue='brand')
sns.scatterplot(data=df,x='hp',y='mpg',ax=axis[1,0],hue='brand')
sns.scatterplot(data=df,x='hp',y='time-to-60',ax=axis[1,1],hue='brand')
plt.show()
```

```python
# 2. Create bar plots for the following, make sure all the plots appear in the same plane with
# x-labels and y-labels evenly spaced.
# a. Create a bar plot that shows the visual representation of hp, mpg, weight,
# time-to-60 with respect to the number of cylinders for each of the brands.
# b. Create a bar plot that shows the visual representation of hp, mpg, weight and
# time-to-60 with respect to the years, for each brand.
```

```python
#a
fig,axis=plt.subplots(2,2,figsize=(16,8))
sns.barplot(data=df,x='cylinders',y='hp',hue='brand',ax=axis[0,0])
sns.barplot(data=df,x='cylinders',y='mpg',hue='brand',ax=axis[0,1])
sns.barplot(data=df,x='cylinders',y='weightlbs',hue='brand',ax=axis[1,0])
sns.barplot(data=df,x='cylinders',y='time-to-60',hue='brand',ax=axis[1,1])
plt.show()
```

```python
#b
fig,axis=plt.subplots(2,2,figsize=(16,8))
sns.barplot(data=df,x='year',y='hp',hue='brand',ax=axis[0,0])
sns.barplot(data=df,x='year',y='mpg',hue='brand',ax=axis[0,1])
sns.barplot(data=df,x='year',y='weightlbs',hue='brand',ax=axis[1,0])
sns.barplot(data=df,x='year',y='time-to-60',hue='brand',ax=axis[1,1])
plt.show()
```

```python
# 6. Create a histogram for the following:
# a. Create histograms for hp, mpg, time-to-60, weight lbs, cubic inches, year, with
# respect to brand.
```

```python
fig,axis=plt.subplots(3,2,figsize=(16,10))
sns.histplot(data=df,x='hp',hue='brand',ax=axis[0,0])
sns.histplot(data=df,x='mpg',hue='brand',ax=axis[0,1])
sns.histplot(data=df,x='time-to-60',hue='brand',ax=axis[1,0])
sns.histplot(data=df,x='weightlbs',hue='brand',ax=axis[1,1])
sns.histplot(data=df,x='cubicinches',hue='brand',ax=axis[2,0])
sns.histplot(data=df,x='year',hue='brand',ax=axis[2,1])
```

```python
# 7. Create line plots for the following:
# a. Line plot between hp-mpg, hp-(time-to-60), mpg-(time-to-60), hp-weight lbs,
# mpg-weight lbs, cubic inches-weight lbs.
```

```python
fig,axis=plt.subplots(3,2,figsize=(16,10))
sns.lineplot(data=df,x='hp',y='mpg',ax=axis[0,0])
sns.lineplot(data=df,x='hp',y='time-to-60',ax=axis[0,1])
sns.lineplot(data=df,x='mpg',y='time-to-60',ax=axis[1,0])
sns.lineplot(data=df,x='hp',y='weightlbs',ax=axis[1,1])
sns.lineplot(data=df,x='mpg',y='weightlbs',ax=axis[2,0])
sns.lineplot(data=df,x='cubicinches',y='weightlbs',ax=axis[2,1])
plt.show()



```
