Q1.Which state is having the highest honey production througout all these years?  represent it using a bar graph.

```python
high_prod_state=df.groupby('state')['production'].sum().sort_values(ascending=False).reset_index()
```

```python
high_prod_state
```

```python
plt.figure(figsize=(10,10))
sns.barplot(data=high_prod_state,x='production',y='state',palette='hsv')
plt.show()
```

Q2.Which state is having the highest and the lowest yield per colony?represent it using bar graph.

```python
high_yield_col=df.groupby('state')['yield_per_colony'].mean().sort_values(ascending=False).reset_index()
```

```python
high_yield_col
```

```python
plt.figure(figsize=(10,10))
sns.barplot(data=high_yield_col,x='yield_per_colony',y='state',palette='hsv')
plt.show()
```

Q3.Which year witnesses the highest profit in honey production?use bar graph to plot it.

```python
high_prof_year=df.groupby('year')['value_of_production'].sum().sort_values(ascending=False).reset_index()
```

```python
high_prof_year
```

```python
plt.figure(figsize=(10,10))
sns.barplot(data=high_prof_year,x='year',y='value_of_production',palette='hsv')
plt.xticks(rotation=90)
plt.show()
```

Q4.Which state was having the highest stock in 2020?represent it using the bar graph.

```python
df_2020=df[df['year']==2020]
```

```python
high_stock_state=df_2020.groupby('state')['stocks'].sum().sort_values(ascending=False).reset_index()
```

```python
high_stock_state
```

```python
plt.figure(figsize=(10,10))
sns.barplot(data=high_stock_state,x='state',y='stocks',palette='hsv')
plt.xticks(rotation=90)
plt.show()
```
