```python
outliers = ['price','minimum_nights','calculated_host_listings_count']
```

```python
for i in outliers:
  q1 = df[i].quantile(0.25)
  q3 = df[i].quantile(0.75)
  iqr = q3 - q1
  ll = q1 - 1.5*iqr
  ul = q3 + 1.5*iqr
  df = df[(df[i]>ll) & (df[i]<ul)]
```

```python
for col in df.select_dtypes(exclude = 'object'):
  plt.boxplot(df[col])
  plt.title(col)
  plt.show()
```

```python
def prices(price):
  if price < 50:
    return 'low'
  elif price > 50 and price < 200:
    return 'medium'
  else:
    return 'high'
```

```python
df['price_range'] = df['price'].apply(prices)
```

```python
df
```

```python
sns.scatterplot(data = df,x='latitude',y='longitude',hue='price_range')
plt.show()
```
